```python
#define Function Here....medianvalues 


def impute_median(series):
    return series.fillna(series.median())
```

```python
df['Rating']=df['Rating'].transform(impute_median)
```

```python
df.isnull().sum()
```

```python
#mode for categorical values


print(df['Type'].mode())
print(df['Current Ver'].mode())
print(df['Android Ver'].mode())
```

```python
df['Type'].fillna(str(df['Type'].mode().values[0]),inplace=True)
df['Current Ver'].fillna(str(df['Current Ver'].mode().values[0]),inplace=True)
df['Android Ver'].fillna(str(df['Android Ver'].mode().values[0]),inplace=True)
```

```python
df.isnull().sum()
```

```python
df.head()
```

```python
df['Reviews']=df['Reviews'].apply(lambda x:float(x))
df['Installs'] = df['Installs'].apply(lambda x: str(x).replace('+','').replace(',','')).astype('int')
df['Price'] = df['Price'].apply(lambda x: str(x).replace('$','')).astype('float')
```

# Data Visualization

```python
rama = df.groupby('Category')
x=rama['Rating'].mean()
y=rama['Price'].sum()
z=rama['Reviews'].max()
print(x)
print(y)
print(z)
```

```python
plt.figure(figsize=(12,6))
plt.plot(x,'ro',color='g');
plt.xticks(rotation=90);
plt.title('Category wise Rating');
plt.xlabel('Categories');
plt.ylabel('Rating-->');
```

```python
plt.figure(figsize=(12,6))
plt.plot(y,'r--',color='b');
plt.xticks(rotation=90);
plt.title('Category wise Rating');
plt.xlabel('Categories');
plt.ylabel('Price-->');
```

```python
plt.figure(figsize=(12,6))
plt.plot(z,'g^',color='b');
plt.xticks(rotation=90);
plt.title('Category wise Rating');
plt.xlabel('Categories');
plt.ylabel('Max-->');
```
