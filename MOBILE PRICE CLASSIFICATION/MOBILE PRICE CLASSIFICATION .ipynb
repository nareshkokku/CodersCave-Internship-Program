```python
A=rama.iloc[: , :-1].values
B=rama.iloc[: , -1].values
```

```python
A[0]
```

```python
B[0]
```

```python
from sklearn.model_selection import train_test_split
```

```python
A_train,A_test,B_train,B_test=train_test_split(A,B,test_size=0.33,random_state=0)
```

```python
A_train.shape
```

```python
A_test.shape
```

```python
from sklearn.preprocessing import StandardScaler
nani= StandardScaler()
A_train=nani.fit_transform(A_train)
A_test=nani.transform(A_test)
```

```python
#Random forest classification 

from sklearn.ensemble import RandomForestClassifier
classifier1=RandomForestClassifier(n_estimators =15, criterion ='entropy')
classifier1.fit(A_train,B_train)
```

```python
B_pred=classifier1.predict(A_test)
```

```python
B_pred
```

```python
from sklearn.metrics import confusion_matrix
ns=confusion_matrix(B_test,B_pred)
sns.heatmap(ns,annot=True)
```

```python
 from sklearn.metrics import accuracy_score
accuracy_score(B_test,B_pred)
```

```python


from sklearn.neighbors import KNeighborsClassifier
classifier=KNeighborsClassifier(n_neighbors=39,metric='minkowski')
classifier.fit(A_train,B_train)
```

```python
B_pred=classifier.predict(A_test)
```

```python
B_pred
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(B_test, B_pred)
```
